## Ejercicio 1

Utilizando una tupla, escribe una función a la que se le pase el número del mes y devuelva su nombre.

Crea un programa principal para probarla.


```python
def nombre_mes(mes: int) -> str:
    """
    Devuelve el nombre del mes correspondiente al número entero dado.

    Args:
        mes (int): El número del mes (entre 1 y 12).

    Returns:
        str: El nombre del mes correspondiente al número dado, o 'Mes incorrecto' si el número no es válido.

    >>> nombre_mes(1)
    'Enero'
    >>> nombre_mes(5)
    'Mayo'
    >>> nombre_mes(12)
    'Diciembre'
    >>> nombre_mes(13)
    'Mes incorrecto'
    >>> nombre_mes(0)
    'Mes incorrecto'
    """
    meses = ('Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto',
             'Septiembre', 'Octubre', 'Noviembre', 'Diciembre')
    if mes < 1 or mes > 12:
        return 'Mes incorrecto'
    else:
        return meses[mes-1]
```

## Ejercicio 2

Busca el algoritmo para calcular la letra del DNI y crea una función que use listas para devolver la letra del mismo.

```python
def letra_dni(dni: int) -> str:
    """
    Devuelve la letra del DNI español correspondiente a un número de DNI.

    Args:
        dni (int): El número de DNI sin la letra (solo los 8 dígitos).

    Returns:
        str: La letra correspondiente al número de DNI según el algoritmo oficial.

    >>> letra_dni(12345678)
    'Z'
    >>> letra_dni(98765432)
    'M'
    >>> letra_dni(45678901)
    'G'
    >>> letra_dni(11111111)
    'H'
    >>> letra_dni(23456789)
    'D'
    """
    letras_dni = ('T', 'R', 'W', 'A', 'G', 'M', 'Y', 'F', 'P', 'D', 
                  'X', 'B', 'N', 'J', 'Z', 'S', 'Q', 'V', 'H', 'L', 
                  'C', 'K', 'E')
    resto = dni % 23
    return letras_dni[resto]
```

# Ejercicio 3

Crea una función a la que se le pase un día de la semana y devuelva el siguiente, por ejemplo, si le pasamos "lunes" nos devolverá "martes".

```python
def siguiente_dia(dia: str) -> str:
    """
    Devuelve el siguiente día de la semana dado un día de la semana.

    Args:
        dia (str): El nombre del día de la semana.

    Returns:
        str: El nombre del siguiente día de la semana.

    >>> siguiente_dia('Lunes')
    'Martes'
    >>> siguiente_dia('Miércoles')
    'Jueves'
    >>> siguiente_dia('Sábado')
    'Domingo'
    >>> siguiente_dia('Domingo')
    'Lunes'
    >>> siguiente_dia('Viernes')
    'Sábado'
    """
    dias = ('Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo')
    posicion = dias.index(dia)
    return dias[(posicion+1) % len(dias)]
```

## Ejercicio 4: AVANZADO

En matemáticas, el número imaginario $i$ es $\sqrt{-1}$. Así sus potencias serán:

$i^0 = 1$

$i^1 = i$

$i^2 = -1$

$i^3 = -i$

$i^4 = 1 $

$i^5 = i $

$...$

Escribe una función a la que se le pase el exponente para una potencia de $i$ y devuelva, como cadena, el resultado.

[Ejercicio en codewars] (https://www.codewars.com/kata/5a97387e5ee396e70a00016d/train/python)

```python
def potencia_i(exp: int) -> str:
    """
    Devuelve el resultado de elevar `i` a la potencia dada, como una cadena.

    La función calcula `i` elevado a una potencia `exp`, donde `i` es la unidad imaginaria. 
    Dado que las potencias de `i` siguen un ciclo periódico de 4, la función retorna el resultado correspondiente.
    
    Los resultados posibles de las potencias de `i` son:
    - `i^0 = 1`
    - `i^1 = i`
    - `i^2 = -1`
    - `i^3 = -i`
    - El ciclo se repite cada 4 términos, por lo que `i^4 = 1`, `i^5 = i`, etc.

    Args:
        exp (int): El exponente al que se eleva la unidad imaginaria `i`.

    Returns:
        str: El resultado de la potencia de `i` como cadena de texto. 
             Puede ser uno de los siguientes valores: "1", "i", "-1", "-i".

    >>> potencia_i(0)
    '1'
    >>> potencia_i(1)
    'i'
    >>> potencia_i(2)
    '-1'
    >>> potencia_i(3)
    '-i'
    >>> potencia_i(4)
    '1'
    >>> potencia_i(5)
    'i'
    >>> potencia_i(10)
    '-1'
    """
    resultados = ("1", "i", "-1", "-i")
    resto = exp % len(resultados)
    return resultados[resto]
```
